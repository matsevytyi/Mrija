```python
import pandas as pd

# Read the Excel file
file_path = '/content/Interests.xlsx'  # Replace with the path to your Excel file
df = pd.read_excel(file_path)

# Extract relevant columns
df_interests = df[['Title', 'Data Value', 'Scale ID']]

# Create separate columns for each RIASEC value
df_interests['R'] = df_interests['Data Value'][0::9].reset_index(drop=True)
df_interests['I'] = df_interests['Data Value'][1::9].reset_index(drop=True)
df_interests['A'] = df_interests['Data Value'][2::9].reset_index(drop=True)
df_interests['S'] = df_interests['Data Value'][3::9].reset_index(drop=True)
df_interests['E'] = df_interests['Data Value'][4::9].reset_index(drop=True)
df_interests['C'] = df_interests['Data Value'][5::9].reset_index(drop=True)

print(df_interests)

# Pivot table to get separate columns for each RIASEC value
pivot_table = df_interests.pivot_table(index='Title', values=['R', 'I', 'A', 'S', 'E', 'C'], aggfunc='first')
print(pivot_table)

# Reset the index to get 'Title' as a regular column
pivot_table = pivot_table.reset_index()

# Extract Occupational Interest High-Point columns
high_point_cols = df[df['Scale ID'] == 'IH'][['Title', 'Data Value']].rename(columns={'Data Value': 'IH'})
high_point_cols['IH_index'] = high_point_cols.groupby('Title').cumcount() + 1
high_point_cols = high_point_cols.pivot_table(index='Title', columns='IH_index', values='IH', aggfunc='first')

# Rename columns for clarity
high_point_cols.columns = [f'IH_{i}_index' for i in range(1, len(high_point_cols.columns) + 1)]

# Merge the two tables on 'Title'
result = pd.merge(pivot_table, high_point_cols, on='Title', how='outer')

# Save the result to a new Excel file
result.to_excel('/content/Final_Interests.xlsx', index=False)

```
